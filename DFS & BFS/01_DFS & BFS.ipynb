### 스택 자료구조
- 먼저 들어온 데이터가 나중에 나감(선입후출)
- 입구와 출구가 동일한 형태 <br>
e.g) Python의 List 자료형에서 append()와 pop() 메소드

```python
stack = []

stack.append(5)
stack.append(2)
stack.append(3)
stack.append(7)
stack.pop() # 맨 마지막 요소 제거
stack.append(1)
stack.append(4)
stack.pop() # parameter에 인덱스를 주면 그 인덱스를 삭제

print(stack[::-1]) # 최상단 원소부터 출력
print(stack) # 최하단 원소부터 출력
```

### 큐 자료구조
- 먼저 들어 온 데이터가 먼저 나감(선입선출, First in First out)
- 입구과 출구가 일자로 뚫려있는 터널과 같은 형태 <br>
e.g) 은행업무 볼 때 순번/대기줄에 따라 보는 것

```python
# Python에서는 queue로 사용할 수 있는 내장된 자료형이 없다 => 라이브러리 사용
from collections import deque
queue = deque()

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
queue.popleft() # 큐를 구현하기 위해 제일 처음 들어온 원소 제거
queue.append(1)
queue.append(4)
queue.popleft()

print(queue) # 먼저 들어온 순서대로 출력
queue.reverse()
print(queue) # 나중에 들어온 순서대로 출력
```

### 재귀 함수(Recursive Function)
- 자기 자신을 다시 호출하는 함수
- stack 대신에 많이 사용됨. 유사한 구조를 가지기도 하고 <br>
why? 재귀함수가 실행될 때 가장 마지막의 재귀함수부터 실행되어야하기 때문

```python
def recursive_function():
    print('재귀 함수를 호출합니다.')
    recursive_function()

recursive_function()
```

```python
# 무한루프를 피하기 위해 재귀함수의 종료 조건을 명시해야 함
def recursive_function(i):
    if i == 100:
        return
    print(f'{i}번째 재귀함수에서 {i+1}번째 재귀함수를 호출합니다.')
    recursive_function(i+1)
    print(f'{i}번째 재귀함수를 호출합니다.')

recursive_function(1)
```

```python
# 반복으로 구현한 팩토리얼
def factorial_iterative(n):
    output = 1
    for i in range(1, n+1):
        output *= i
    return output
# 재귀적으로 구현한 팩토리얼 <=> 점화식
def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n-1)
```

```python
%%time
factorial_iterative(30)
```

```python
%%time
factorial_recursive(30)
```

## BFS(Breadth-First Search), 너비우선탐색
- 그래프에서 가장 가까운 부분을 우선적으로 탐색
- queue 자료구조를 이용 <br>

1. 탐색 시작 노드를 큐에 삽입하고 방문 처리 <br>
2. 큐에서 노드를 꺼낸 뒤에 해당 노드의 인접 노드 중에서 방문하지 않은 노드를 모두 큐에 삽입하고 방문 처리 <br>
3. 더 이상 2번 과정을 수행할 수 없을 때까지 반복  <br>

1을 먼저 넣고. 1을 뺀 뒤에
1에서 가장 가까운 노드: 2, 3, 8
얘들을 2, 3, 8 이렇게 한꺼번에 큐에 넣음 

그 다음 2 꺼냄. 글고 꺼낸 2와 연결된 놈에서  방문하지 않은 7을 꺼내 큐에 넣음


```python
from collections import deque

# BFS 메소드
def bfs(graph, start, visited):
    # 큐(queue) 구현을 위해 deque 라이브러리 사용
    # 시작(맨 처음)/ 현재 노드 방문처리
    queue = deque([start]) # => 이따가 그래프를 중첩리스트로 정의해서 그럼


    visited[start] = True
    # queue가 빌 때까지 반복
    while queue:
        # 큐에서 하나의 원소를 뽑아 출력
        v = queue.popleft()
        print(v, end = ' ')
        # 인접 노드와 같은, 방문하지 않은 인접 원소들을 큐에 넣음
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
            
```

```python
graph = [
    [],
    [2,3,8], 
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

visited = [False] * 9
dfs(graph, 1, visited)
```

2. 방문 지점에서 상하좌우를 살펴보며 방문을 진행하다보면 연결된 모든 지점 방문 가능
3. 모든 노드에 대해 1,2번 과정을 반복하며 방문하지 않은 지점 수를 카운트한다.

```python
# DFS로 특정 노드 방문하고 연결된 모든 노드 방문
def dfs(x,y):
    # 주어진 범위를 벗어나면 바로 종료
    if x <= -1 or x >=n or y <=1 or y >= m :
        return False
    if graph[x][y] == 0:
        # 방문처리. 나중엔 다시 안 오겠지
        graph[x][y] = 1
 
        dfs(x - 1,y)
        dfs(x,y - 1)
        dfs(x + 1,y)
        dfs(x,y + 1)
        return True
    # 애는 모든 함수가 출력이 완료되면 False가 나옴
    return False 

n,m = map(int, input('행, 열을 순서대로 spacebar로 구분하여 입력:').split())

graph = []
for i in range(n):
    graph.append( list(map(int,input(f'{i}번째 행 요소 spacebar로 구분해 입력: ').split(' ') ) ) )

print(f'graph: {graph}')

# 패딩이 안되서 계속 0이 나오나?
result = 0
for i in range(n):
    for j in range(m):
        print(f'{i}행 {j}열 그래프{graph[i][j]} 결과: {dfs(i,j)}' )
        if dfs(i,j) == True:
            result += 1
    
print(result)
```
