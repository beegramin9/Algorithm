### 스택 자료구조
- 먼저 들어온 데이터가 나중에 나감(선입후출)
- 입구와 출구가 동일한 형태 <br>
e.g) Python의 List 자료형에서 append()와 pop() 메소드

```python
stack = []

stack.append(5)
stack.append(2)
stack.append(3)
stack.append(7)
stack.pop() # 맨 마지막 요소 제거
stack.append(1)
stack.append(4)
stack.pop() # parameter에 인덱스를 주면 그 인덱스를 삭제

print(stack[::-1]) # 최상단 원소부터 출력
print(stack) # 최하단 원소부터 출력
```

### 큐 자료구조
- 먼저 들어 온 데이터가 먼저 나감(선입선출, First in First out)
- 입구과 출구가 일자로 뚫려있는 터널과 같은 형태 <br>
e.g) 은행업무 볼 때 순번/대기줄에 따라 보는 것

```python
# Python에서는 queue로 사용할 수 있는 내장된 자료형이 없다 => 라이브러리 사용
from collections import deque
queue = deque()

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
queue.popleft() # 큐를 구현하기 위해 제일 처음 들어온 원소 제거
queue.append(1)
queue.append(4)
queue.popleft()

print(queue) # 먼저 들어온 순서대로 출력
queue.reverse()
print(queue) # 나중에 들어온 순서대로 출력
```

### 재귀 함수(Recursive Function)
- 자기 자신을 다시 호출하는 함수
- stack 대신에 많이 사용됨. 유사한 구조를 가지기도 하고 <br>
why? 재귀함수가 실행될 때 가장 마지막의 재귀함수부터 실행되어야하기 때문

```python
def recursive_function():
    print('재귀 함수를 호출합니다.')
    recursive_function()

recursive_function()
```

```python
# 무한루프를 피하기 위해 재귀함수의 종료 조건을 명시해야 함
def recursive_function(i):
    if i == 100:
        return
    print(f'{i}번째 재귀함수에서 {i+1}번째 재귀함수를 호출합니다.')
    recursive_function(i+1)
    print(f'{i}번째 재귀함수를 호출합니다.')

recursive_function(1)
```

```python
# 반복으로 구현한 팩토리얼
def factorial_iterative(n):
    output = 1
    for i in range(1, n+1):
        output *= i
    return output
# 재귀적으로 구현한 팩토리얼 <=> 점화식
def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n-1)
```

```python
%%time
factorial_iterative(30)
```

```python
%%time
factorial_recursive(30)
```
